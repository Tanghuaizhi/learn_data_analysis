# 一、收集数据

```python
import pandas as pd
import numpy as np

import requests
import os
import json

% matplotlib inline
import matplotlib.pyplot as plt

# 解决内容显示不全的问题
pd.set_option('max_colwidth', 200)
```

**1、导入给定的已下载WeRateDogs推特数据，形成DataFrame数据`twitters`**

```python
twitters = pd.read_csv('twitter-archive-enhanced.csv')
```

2、下载给定网址的推特图像预测数据

```python
url = 'https://raw.githubusercontent.com/udacity/new-dand-advanced-china/master/%E6%95%B0%E6%8D%AE%E6%B8%85%E6%B4%97/WeRateDogs%E9%A1%B9%E7%9B%AE/image-predictions.tsv'
response = requests.get(url)
with open(url.split('/')[-1], mode='wb') as file:
    file.write(response.content)    
```

**导入推荐图像预测数据，形成DataFrame数据`images`**

```python
images = pd.read_csv('image-predictions.tsv', sep='\t')
```

3、导入项目提供的tweet_json.txt文件  
读入tweet ID、retweet_count 和 favorite_count 字段，形成DataFrame数据`twitter_additions`
并保存为`twitter_additions.csv`

```python
twitter_additions_list = []
with open('tweet_json.txt', encoding='utf-8', mode='r') as file:
    for line in file:    
        parsed_json = json.loads(line)
        twitter_additions_list.append({'tweet_ID':parsed_json['id'],
                                       'retweet_count':parsed_json['retweet_count'],
                                       'favorite_count':parsed_json['favorite_count']})
twitter_additions = pd.DataFrame(twitter_additions_list, columns = ['tweet_ID', 'retweet_count', 'favorite_count'])
twitter_additions.to_csv('twitter_additions.csv')
```

**重新导入`twitter_additions.csv`**

```python
twitter_additions = pd.read_csv('twitter_additions.csv')
```

# 二、评估数据

**关键要点**  

清洗这个项目的数据时要牢记几个要点：  
- 我们只需要含有图片的原始评级 (不包括转发)。尽管数据集中有 5000 多条数据，但是并不是所有都是狗狗评分，并且其中有一些是转发。
- 完整地评估和清理整个数据集将需要大量时间，实践和展示数据处理技巧没有必要将这个数据集全部清理。因此，本项目的要求只是评估和清理此数据集中的至少 8 个质量问题和至少 2 个整洁度问题。
- 根据 整洁数据 tidy data 的规则要求，本项目的数据清理应该包括将三个数据片段进行合并。
- 如果分子评级超过分母评级，不需要进行清洗。这个 特殊评分系统 是 WeRateDogs 人气度较高的主要原因。（同样，也不需要删除分子小于分母的数据）
- 不必收集 2017 年 8 月 1 日之后的数据，你可以收集到这些推特的基本信息，但是你不能收集到这些推特对应的图像预测数据，因为你没有图像预测算法的使用权限。

## （三）评估结果：问题记录
### 1. 数据质量
**`twitters` 表格**
- id类的列应是字符串，不是数字（tweet_id、in_reply_to_status_id、in_reply_to_user_id、retweeted_status_id、retweeted_status_user_id）
- timestamp 中的后6位即' +0000'没有意义
- source应为iphone等，其它标签意义不大
- text正文应和短网址分开，拆分为正文和网址两列
- retweeted_status_id、retweeted_status_user_id、retweeted_status_timestamp数据大范围缺失**（无法清理）**
- expanded_urls小部分数据缺失**（无法清理）**
- rating_numerator个别值远超过10，如960，有的自动提取文字时出现错误，如911；文字提取错误的情况也出现在rating_denominator、name中
- rating_numerator出现无效分（0分）
- name、doggo、floofer、pupper、puppo中存在空值（NaN），原因是个别行从retweeted_status_id开始出现“错列”填写情况，向左1-3列或向右1列，至少98个

**`images` 表格**
- P1、P2、P3的字符中字母大小写不统一
- 最大概率P1为False，不属于狗
- p1_conf、p2_conf、p3_conf位数保留不一致，保留到小数点后2位即可
- 并没有p4以及相关的列，即img_num列中的4没有意义，检验后若按最大概率，全部可取1

**`twitter_additions` 表格**
- tweet_ID应是字符串，不是数字
- favorite_count部分为0**（无法清理）**

### 2. 数据结构
- twitter_additions、images应和twitters表格合并
- `twitters` 表的text列中的正文内容应和短网址分开，拆分为正文和网址两列
- doggo、floofer、pupper、puppo含义一致，应合并为一列stage

以上共包含18个问题，对其中12个质量问题、3个结构问题进行处理。

```python
twitters_copy = twitters.copy()
images_copy = images.copy()
twitter_additions_copy = twitter_additions.copy()
```

## （一）结构问题：补充表images、twitter_additions到表twitters中  
### 1. 定义
- 用pandas的[表合并](https://pandas.pydata.org/pandas-docs/stable/merging.html)操作，按tweet_id将表`images`、`twitter_additions`与当前主表`twitters`合并，形成新主表`df_master`
- 将id类列转换为str格式
- 导出主表，保存为`twitter_archive_master.csv`  

### 2. 代码  

```python
twitters_copy.set_index('tweet_id', inplace=True)
images_copy.set_index('tweet_id',  inplace=True)
twitter_additions_copy.set_index('tweet_ID', inplace=True)
```

```python
df_master = pd.concat([twitters_copy, images_copy, twitter_additions_copy], axis=1, join='inner')
df_master.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
df_master.index = df_master.index.astype('str')
```

```python
id_col_names = ['in_reply_to_user_id', 'in_reply_to_user_id', 'retweeted_status_id', 'retweeted_status_user_id','in_reply_to_status_id']
df_master.loc[:, id_col_names] = df_master.loc[:, id_col_names].astype('str')
```

```python
# 将表中所有的None值替换为NaN
df_master.replace('None', np.nan, inplace=True)
```

### 3. 测试

```python
df_master.info()
```

```python
# 检查主表是否有重复
df_master[df_master.duplicated()]
```

## （二）结构问题：text列中的正文内容应和短网址分开，拆分为正文和网址两列  
### 1. 定义
- 用正则表达式和pandas的`str.extract`方法，从表df_master的列*text*中提取提取网址，保存为新列twitter_urls
- 用pandas的`str.split`方法，从列text中去掉网址

### 2. 代码  

```python
import re
```

```python
df_master['twitter_urls'] = df_master.text.str.extract(r'(https:.*[A-Z])' , expand=True)
```

```python
df_master['text'] = df_master.text.str.split('https://', expand=True)[0]
```

## （三）结构问题：doggo、floofer、pupper、puppo应合并为一列stage
### 1.定义
- 用`str.cat`方法从主表中提取列*doggo、floofer、pupper、puppo*的值，合并形成新的列*stage*  
- 删除列*doggo、floofer、pupper、puppo*  

### 2.代码

```python
df_master['stage'] = df_master['doggo'].str.cat([df_master['floofer'], df_master['pupper'], df_master['puppo']], na_rep='')
```

```python
df_master.drop(columns=['doggo', 'floofer', 'pupper', 'puppo'], inplace=True)
```

## （八）一致性问题
### 1.定义
- 用str方法处理timestamp，去掉后6位即' +0000'
- 用pandas和正则表达式将列source提取出iphone等类型值
- 将P1、P2、P3字符中的字母统一为小写
- 将p1_conf、p2_conf、p3_conf位数全部保留到小数点后2位  

以上均采用pandas的str[相关方法](https://pandas.pydata.org/pandas-docs/stable/text.html#splitting-and-replacing-strings)处理

### 2.代码

```python
# 处理列timestamp
df_master.timestamp = df_master.timestamp.str.split(' \+').str[0]
```

```python
# 处理列source
df_master.source = df_master.source.str.extract(r'<.*>(.*)<.*>' , expand=True)
```

```python
# 处理列P1、P2、P3、stage
list_col_lower = ['p1','p2','p3', 'stage']
for a in list_col_lower:
    df_master.loc[:, a] = df_master.loc[:, a].str.lower()
```

```python
# 处理列p1_conf、p2_conf、p3_conf
list_col_round = ['p1_conf', 'p2_conf', 'p3_conf']
for b in list_col_round:
    df_master.loc[:, b] = df_master.loc[:, b].round(decimals=2)
```
