# 一、收集数据

```python
import pandas as pd
import numpy as np

import requests
import os
import json

% matplotlib inline
import matplotlib.pyplot as plt

# 解决内容显示不全的问题
pd.set_option('max_colwidth', 200)
```

**1、导入给定的已下载WeRateDogs推特数据，形成DataFrame数据`twitters`**

```python
twitters = pd.read_csv('twitter-archive-enhanced.csv')
```

2、下载给定网址的推特图像预测数据

```python
url = 'https://raw.githubusercontent.com/udacity/new-dand-advanced-china/master/%E6%95%B0%E6%8D%AE%E6%B8%85%E6%B4%97/WeRateDogs%E9%A1%B9%E7%9B%AE/image-predictions.tsv'
response = requests.get(url)
with open(url.split('/')[-1], mode='wb') as file:
    file.write(response.content)    
```

**导入推荐图像预测数据，形成DataFrame数据`images`**

```python
images = pd.read_csv('image-predictions.tsv', sep='\t')
```

3、导入项目提供的tweet_json.txt文件  
读入tweet ID、retweet_count 和 favorite_count 字段，形成DataFrame数据`twitter_additions`
并保存为`twitter_additions.csv`

```python
twitter_additions_list = []
with open('tweet_json.txt', encoding='utf-8', mode='r') as file:
    for line in file:    
        parsed_json = json.loads(line)
        twitter_additions_list.append({'tweet_ID':parsed_json['id'],
                                       'retweet_count':parsed_json['retweet_count'],
                                       'favorite_count':parsed_json['favorite_count']})
twitter_additions = pd.DataFrame(twitter_additions_list, columns = ['tweet_ID', 'retweet_count', 'favorite_count'])
twitter_additions.to_csv('twitter_additions.csv', index = False)
```

## （一）结构问题：补充表images、twitter_additions到表twitters中  
### 1. 定义
- 用pandas的[表合并](https://pandas.pydata.org/pandas-docs/stable/merging.html)操作，按tweet_id将表`images`、`twitter_additions`与当前主表`twitters`合并，形成新主表`df_master`
- 将id类列转换为str格式


### 2. 代码  

```python
twitters_copy.set_index('tweet_id', inplace=True)
images_copy.set_index('tweet_id',  inplace=True)
twitter_additions_copy.set_index('tweet_ID', inplace=True)
```

```python
df_master = pd.concat([twitters_copy, images_copy, twitter_additions_copy], axis=1, join='inner')
```

```python
df_master.index = df_master.index.astype('str')
```

```python
id_col_names = ['in_reply_to_user_id', 'in_reply_to_user_id', 'retweeted_status_id', 'retweeted_status_user_id','in_reply_to_status_id']
df_master.loc[:, id_col_names] = df_master.loc[:, id_col_names].astype('str')
```

```python
# 将表中所有的None值替换为NaN
df_master.replace('None', np.nan, inplace=True)
```

### 3. 测试

```python
df_master.info()
```

```python
# 检查主表是否有重复
df_master[df_master.duplicated()]
```

## （二）结构问题：text列中的正文内容应和短网址分开，拆分为正文和网址两列

- 用正则表达式和pandas的`str.extract`方法，从表df_master的列*text*中提取提取网址，保存为新列twitter_urls
- 用pandas的`str.split`方法，从列text中去掉网址

### 2. 代码  

```python
import re
```

```python
df_master['twitter_urls'] = df_master.text.str.extract(r'(https:.*[A-Z])' , expand=True)
```

```python
df_master['text'] = df_master.text.str.split('https://', expand=True)[0]
```

## （八）一致性问题
### 1.定义
- 用str方法处理timestamp，去掉后6位即' +0000'
- 用pandas和正则表达式将列source提取出iphone等类型值
- 将P1、P2、P3字符中的字母统一为小写
- 将p1_conf、p2_conf、p3_conf位数全部保留到小数点后2位  

以上均采用pandas的str[相关方法](https://pandas.pydata.org/pandas-docs/stable/text.html#splitting-and-replacing-strings)处理

### 2.代码

```python
# 处理列timestamp
df_master.timestamp = df_master.timestamp.str.split(' \+').str[0]
```

```python
# 处理列source
df_master.source = df_master.source.str.extract(r'<.*>(.*)<.*>' , expand=True)
```

```python
# 处理列P1、P2、P3、stage
list_col_lower = ['p1','p2','p3', 'stage']
for a in list_col_lower:
    df_master.loc[:, a] = df_master.loc[:, a].str.lower()
```

```python
# 处理列p1_conf、p2_conf、p3_conf
list_col_round = ['p1_conf', 'p2_conf', 'p3_conf']
for b in list_col_round:
    df_master.loc[:, b] = df_master.loc[:, b].round(decimals=2)
```
